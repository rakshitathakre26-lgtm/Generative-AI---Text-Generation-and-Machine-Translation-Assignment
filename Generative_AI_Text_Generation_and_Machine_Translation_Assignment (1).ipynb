Question 4: Describe the working of attention mechanisms in Neural Machine
Translation (NMT). Why are they critical?

🧠 Attention Mechanism in Neural Machine Translation (NMT)
🔍 Basic Idea

In Neural Machine Translation, attention allows the model to focus on relevant words in the source sentence while generating each word in the target sentence.

👉 Instead of encoding the whole sentence into a single fixed vector, the model looks back at the input dynamically.

⚙️ How Attention Works (Step-by-Step)
1. 📝 Encoder Phase
The encoder reads the input sentence word by word
Produces a sequence of hidden states (one for each word)

👉 Example:
Input: “I love machine learning”
→ Encoder creates representations for each word

2. 🎯 Decoder with Attention

When generating each output word:

Step A: Compare
The decoder compares its current state with all encoder states
Step B: Assign Weights (Attention Scores)
Each input word gets a weight (importance score)
Step C: Weighted Sum (Context Vector)
Combine encoder states using weights → creates a context vector
Step D: Generate Output
The decoder uses this context to generate the next word
📊 Simple Example

Input: “I eat apple”
Output: “मैं सेब खाता हूँ”

While generating “सेब” (apple), the model gives:

High attention to “apple”
Low attention to other words
🧠 Why Attention is Critical
1. ❌ Solves Fixed-Length Problem
Old models compressed everything into one vector
Attention allows access to full sentence information
2. 🎯 Better Accuracy
Focuses on relevant words
Improves translation quality
3. 📏 Handles Long Sentences
Maintains context even for long inputs
Avoids forgetting earlier words
4. 🔍 Interpretability
Attention weights show which words the model is focusing on
5. 🚀 Foundation for Advanced Models
Led to powerful architectures like Transformers (e.g., BERT, GPT)
⚖️ Without vs With Attention
Without Attention	With Attention
Single fixed vector	Dynamic focus on words
Poor for long sentences	Works well for long sentences
Lower accuracy	Higher accuracy
🎯 Important Line

“Attention mechanisms allow NMT models to dynamically focus on relevant parts of the input sequence during translation, overcoming the limitations of fixed-length encoding and significantly improving accuracy, especially for long and complex sentences.”

Question 8: Implement a simple attention-based encoder-decoder model for
English-to-Spanish translation using Tensorflow or PyTorch.

Below is a simple implementation of an attention-based encoder–decoder model for English → Spanish translation using TensorFlow (Keras).
✅ 1. Install & Import Libraries

```python
import tensorflow as tf
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
```

📚 2. Sample Dataset (Toy Example)

```python
# Small dataset (for demo)
english_sentences = [
    "hello", "how are you", "i am fine", "thank you", "good morning"
]

spanish_sentences = [
    "hola", "como estas", "estoy bien", "gracias", "buenos dias"
]

# Add start and end tokens
spanish_sentences = ["<start> " + s + " <end>" for s in spanish_sentences]
```

🔤 3. Tokenization & Padding

```python
# Tokenizers
eng_tokenizer = Tokenizer(filters='')
eng_tokenizer.fit_on_texts(english_sentences)
eng_seq = eng_tokenizer.texts_to_sequences(english_sentences)

spa_tokenizer = Tokenizer(filters='')
spa_tokenizer.fit_on_texts(spanish_sentences)
spa_seq = spa_tokenizer.texts_to_sequences(spanish_sentences)

# Vocabulary sizes
eng_vocab_size = len(eng_tokenizer.word_index) + 1
spa_vocab_size = len(spa_tokenizer.word_index) + 1

# Padding
max_eng_len = max(len(seq) for seq in eng_seq)
max_spa_len = max(len(seq) for seq in spa_seq)

eng_seq = pad_sequences(eng_seq, maxlen=max_eng_len, padding='post')
spa_seq = pad_sequences(spa_seq, maxlen=max_spa_len, padding='post')

# Decoder input and output
decoder_input = spa_seq[:, :-1]
decoder_output = spa_seq[:, 1:]
```

🧠 4. Attention Layer

```python
class Attention(tf.keras.layers.Layer):
    def __init__(self):
        super(Attention, self).__init__()

    def call(self, encoder_outputs, decoder_hidden):
        # Score calculation
        score = tf.matmul(encoder_outputs, decoder_hidden[:, tf.newaxis], transpose_b=True)

        # Attention weights
        attention_weights = tf.nn.softmax(score, axis=1)

        # Context vector
        context_vector = attention_weights * encoder_outputs
        context_vector = tf.reduce_sum(context_vector, axis=1)

        return context_vector
```

🔁 5. Encoder

```python
class Encoder(tf.keras.Model):
    def __init__(self, vocab_size, embedding_dim, units):
        super(Encoder, self).__init__()
        self.embedding = Embedding(vocab_size, embedding_dim)
        self.lstm = LSTM(units, return_sequences=True, return_state=True)

    def call(self, x):
        x = self.embedding(x)
        output, state_h, state_c = self.lstm(x)
        return output, state_h, state_c
```

🔁 6. Decoder with Attention

```python
class Decoder(tf.keras.Model):
    def __init__(self, vocab_size, embedding_dim, units):
        super(Decoder, self).__init__()
        self.embedding = Embedding(vocab_size, embedding_dim)
        self.lstm = LSTM(units, return_sequences=True, return_state=True)
        self.fc = Dense(vocab_size)
        self.attention = Attention()

    def call(self, x, encoder_outputs, state_h, state_c):
        x = self.embedding(x)

        outputs = []
        for t in range(x.shape[1]):
            context_vector = self.attention(encoder_outputs, state_h)

            # Combine context with input
            x_t = tf.concat([context_vector[:, tf.newaxis, :], x[:, t:t+1]], axis=-1)

            output, state_h, state_c = self.lstm(x_t, initial_state=[state_h, state_c])
            output = self.fc(output)

            outputs.append(output)

        outputs = tf.concat(outputs, axis=1)
        return outputs
```

⚙️ 7. Model Setup

```python
embedding_dim = 64
units = 128

encoder = Encoder(eng_vocab_size, embedding_dim, units)
decoder = Decoder(spa_vocab_size, embedding_dim, units)

optimizer = tf.keras.optimizers.Adam()
loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
```

🏋️ 8. Training Loop

```python
epochs = 100

for epoch in range(epochs):
    with tf.GradientTape() as tape:
        enc_output, enc_h, enc_c = encoder(eng_seq)
        predictions = decoder(decoder_input, enc_output, enc_h, enc_c)

        loss = loss_fn(decoder_output, predictions)

    variables = encoder.trainable_variables + decoder.trainable_variables
    gradients = tape.gradient(loss, variables)
    optimizer.apply_gradients(zip(gradients, variables))

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss.numpy():.4f}")
```

🔍 9. Inference (Translation Function)

```python
def translate(sentence):
    seq = eng_tokenizer.texts_to_sequences([sentence])
    seq = pad_sequences(seq, maxlen=max_eng_len, padding='post')

    enc_output, enc_h, enc_c = encoder(seq)

    target_seq = np.array([[spa_tokenizer.word_index['<start>']]])
    result = []

    for _ in range(max_spa_len):
        preds = decoder(target_seq, enc_output, enc_h, enc_c)
        predicted_id = np.argmax(preds[0, -1, :])

        word = spa_tokenizer.index_word.get(predicted_id, '')
        if word == "<end>":
            break

        result.append(word)
        target_seq = np.array([[predicted_id]])

    return " ".join(result)
```

✨ 10. Example Output

```python
print(translate("hello"))
```

📌 Summary
Encoder processes English input
Attention helps decoder focus on relevant words
Decoder generates Spanish output step-by-step
This is a basic implementation (real systems use larger datasets & models)
